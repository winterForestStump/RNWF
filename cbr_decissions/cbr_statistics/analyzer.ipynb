```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

file_path = "data\\merged_dataset.xlsx"
df_raw = pd.read_excel(file_path, header=0)
```

```python
dates = pd.to_datetime(df_raw.iloc[0, 1:].index, format = "%d/%m/%Y")
categories = df_raw.iloc[0:, 0].values
data = df_raw.iloc[0:, 1:]
data.index = categories
data = data.apply(pd.to_numeric, errors='coerce')
data.fillna(0, inplace=True)
```

```python
data_part = data.iloc[:,60:78]
```

```python
data_part
```

```python
# ADF test function
def adf_test(series, name=""):
    result = adfuller(series)
    print(f"\nADF Test: {name}")
    print(f"ADF statistic: {result[0]:.4f}")
    print(f"p-value: {result[1]:.4f}")
    print("Critical values:")
    for k, v in result[4].items():
        print(f"   {k}: {v:.4f}")
```

```python
for i, sec_type in enumerate(data.index):

    series = data.iloc[i]

    diff_series = series.diff()

    plt.figure(figsize=(10,4))
    plt.plot(series.index, series.values)
    plt.title(f"{sec_type} - Original")
    plt.show()

    plt.figure(figsize=(10,4))
    plt.plot(diff_series.index, diff_series.values)
    plt.title(f"{sec_type} - First Difference")
    plt.show()

    try:
        adf_test(series, f"{sec_type} (original)")
        adf_test(diff_series, f"{sec_type} (first difference)")
    except Exception as e:
        print(f"The problem with running the test {e}")
```

```python
diff_data = data_part.T.diff()
diff_data.dropna(how='all', inplace=True)
# Correlation with all other columns
df_transposed = diff_data
target_series = df_transposed['RUB_CNY_FUND_BONDS']
correlations = df_transposed.drop(columns=['RUB_CNY_FUND_BONDS', 'CNY_FUND_BONDS']).corrwith(target_series)
print("Highest correlation (most similar movement):")
print(correlations.sort_values(ascending=False).head(5))
```
